```python
import pydicom
import cv2
import numpy as np
import pandas as pd
import csv
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
```

```python
def CLAHE(img):
    equ = cv2.equalizeHist(img)
    clahe = cv2.createCLAHE(clipLimit =3.0, tileGridSize=(4,4))
    cl_img = clahe.apply(img)
    ret, thresh3 = cv2.threshold(cl_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    return cl_img
```

```python
test_pic = "../../datasets/image/clean_png/1.2.826.0.1.3680043.8.498.10000118454892828674141713285403927285.dcm.png"
new_path = "../../datasets/image/clean_png/after_clahe.png"


# Read the PNG file
img = cv2.imread(test_pic,0)

# Display the image with the 'gray' colormap
plt.imshow(img, cmap='gray')
plt.axis('off')  # Turn off axis labels
plt.show()

cl_img = CLAHE(img)
cv2.imwrite(new_path, cl_img)
plt.imshow(cl_img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
def CLAHE_WINDOW(img, limit):
    equ = cv2.equalizeHist(img)
    clahe = cv2.createCLAHE(clipLimit = limit, tileGridSize=(4,4))
    cl_img = clahe.apply(img)
    ret, thresh3 = cv2.threshold(cl_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    return cl_img
```

```python
test_pic_window = "../../datasets/image/clean_png/Mass-Training_P_00026_LEFT_MLO_1-1-1.png"


# Read the PNG file
img_wind = cv2.imread(test_pic_window, 0)

# Display the image with the 'gray' colormap
# cv2.imshow("g",img_wind)
# cv2.waitKey(0)

cl_img_wind = CLAHE_WINDOW(img_wind, 4)
cv2.imshow("g",cl_img_wind)
cv2.waitKey(2)

cl_img_wind = CLAHE_WINDOW(img_wind, 5)
cv2.imshow("g",cl_img_wind)
cv2.waitKey(2)

cl_img_wind = CLAHE_WINDOW(img_wind, 10)
cv2.imshow("g",cl_img_wind)
cv2.waitKey(2)
```
